IMPORTANDO BIBLIOTECAS

```python
import csv
import json
```

DEFININDO OS CAMINHOS DOS ARQUIVOS DA FUSAO

```python
pathA = "../data_raw/dados_empresaA.json"
pathB = "../data_raw/dados_empresaB.csv"
```

ETAPA DE EXTRACT - 
ABRINDO OS ARQUIVOS E ARMAZENANDO DADOS 

```python
with open(pathA, "r") as file:
    dados_empresa_A = json.load(file)

quantidade_dados_A = len(dados_empresa_A)
quantidade_dados_A
```

```python
dados_empresa_B = []

with open(pathB, "r", encoding="utf-8") as file:
    spamreader = csv.DictReader(file)
    for row in spamreader:
        dados_empresa_B.append(row)

quantidade_dados_B = len(dados_empresa_B)
quantidade_dados_B
```

```python
quantidade_dados_juncao = quantidade_dados_A + quantidade_dados_B
quantidade_dados_juncao
```

ETAPA TRANSFORM - 
ALTERANDO O NOME DAS CHAVES DO DICIONARIO PARA TODAS SEREM A MESMA 

```python
keysA = dados_empresa_A[0].keys()
keysB = dados_empresa_B[0].keys()
```

```python
keysA
```

```python
keysB
```

```python
key_renamer = {
    "Nome do Item": "Nome do Produto",
    "Classificação do Produto": "Categoria do Produto",
    "Valor em Reais (R$)": "Preço do Produto (R$)",
    "Quantidade em Estoque": "Quantidade em Estoque",
    "Nome da Loja": "Filial",
    "Data da Venda": "Data da Venda"
}

# "CHAVE QUE QUER ALTERAR": "VALOR NOVO PARA CHAVE"
```

```python
dados_empresa_A
# ja tem as chaves corretas

novos_dados_empresa_B = []

for old_dados_B in dados_empresa_B:
    dict_temp = {}
    for old_key, value in old_dados_B.items():
        dict_temp[key_renamer[old_key]] = value
        novos_dados_empresa_B.append(dict_temp)
```

COMBINANDO OS DADOS DAS 2 EMPRESAS

```python
dados_fusao = []

dados_fusao.extend(dados_empresa_A)
dados_fusao.extend(novos_dados_empresa_B)
```

```python
dados_fusao[-1]
```

TRANSFORMANDO TUDO EM LISTA/TABELA PRA EXPORTAR EM CSV

```python
nomes_colunas = list(dados_fusao[-1].keys())

nomes_colunas
```

ETAPA DE LOAD - ENVIANDO TODOS OS DADOS PARA UM LOCAL (ARQUIVO)

```python
path_dados_fusao = "../data_processed/dados_fusao.csv"
```

```python
dados_fusao_lista = [nomes_colunas]

for row in dados_fusao: #armazena em row um dicionario contido na lista da fusao das empresas 
    linha = []
    for column in nomes_colunas: #armazena em column temporariamente o nome de uma string da lista nomes_colunas que sao as colunas para o arquivo csv quie vai ser gerado
        linha.append(row.get(column, "Indisponivel")) #retorna o value associada a column (que sao os nomes das keys do dict que viraram colunas em csv) ou se nao achar retorna indisponivel
    dados_fusao_lista.append(linha)
```

```python
with open(path_dados_fusao, "w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file, quoting=csv.QUOTE_ALL)
    writer.writerows(dados_fusao_lista)
```
